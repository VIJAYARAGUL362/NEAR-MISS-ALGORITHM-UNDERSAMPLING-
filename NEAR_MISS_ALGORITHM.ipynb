IMPORTING THE LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns
```


IMPORTING  THE DATASET

```python
dataset = {
    "ANIMAL":[0,0,0,1,0,1],  # "ANIMAL":["DOG","DOG","DOG","CAT","DOG","CAT"], I inputed the number because then it will make it complicated.
    "HEIGHT":[1,2,2.5,0.6,1.5,0.9],
    "WEIGHT":[5,10,15,2.5,7,1.5]
}

dataset = pd.DataFrame(dataset)
print(dataset)
x=dataset.iloc[:,1:]
y=dataset.iloc[:,0]
```

VISUALIZE THE DATASET

```python
count = Counter(dataset['ANIMAL'])
print(count)

labels = count.keys()
size = count.values()

plt.figure(figsize=(8, 6))
plt.pie(size,labels=labels, autopct='%1.1f%%', startangle=140)
# plt.title('Distribution of Animals')
plt.axis('equal')
plt.show()
```

SCATTER PLOT

```python
sns.scatterplot(x='WEIGHT',y='HEIGHT',data=dataset,hue='ANIMAL')
plt.grid(True)
plt.show()
```

```python
# version 1
from imblearn.under_sampling import NearMiss
nm = NearMiss(version=1,n_neighbors=1)
X_res, y_res = nm.fit_resample(x, y)
print(X_res)
print(y_res)
```

```python
# version 2
from imblearn.under_sampling import NearMiss
nm = NearMiss(version=2,n_neighbors=1)
X_res, y_res = nm.fit_resample(x, y)
print(X_res)
print(y_res)
```

```python
# version 3
from imblearn.under_sampling import NearMiss
nm = NearMiss(version=3,n_neighbors=1,n_neighbors_ver3=4)
X_res, y_res = nm.fit_resample(x, y)
print(X_res)
print(y_res)
```
